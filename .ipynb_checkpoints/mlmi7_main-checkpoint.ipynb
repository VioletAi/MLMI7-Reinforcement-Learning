# policy iteration

```python
from cgitb import small
from typing import Callable
from matplotlib.pyplot import grid
from tqdm import tqdm
import numpy as np

from model import Model, Actions


def policy_iteration(model: Model, maxit: int = 100):

    V = np.zeros((model.num_states,))
    pi = np.zeros((model.num_states,))

    def compute_value(s, a, reward: Callable):
        return np.sum(
            [
                model.transition_probability(s, s_, a)
                * (reward(s, a) + model.gamma * V[s_])
                for s_ in model.states
            ]
        )

    def policy_evaluation():
        for s in model.states:
            R = model.reward(s, pi[s])
            V[s] = compute_value(s, pi[s], lambda *_: R)

    def policy_improvement():
        for s in model.states:
            action_index = np.argmax(
                [compute_value(s, a, model.reward) for a in Actions]
            )
            pi[s] = Actions(action_index)

    for i in tqdm(range(maxit)):
        for _ in range(5):
            policy_evaluation()
        pi_old = np.copy(pi)
        policy_improvement()
        if all(pi_old == pi):
            print("breaking")
            break

    return V, pi


if __name__ == "__main__":
    import matplotlib.pyplot as plt
    from world_config import cliff_world, small_world, grid_world
    from plot_vp import plot_vp

    model = Model(small_world)
    V, pi = policy_iteration(model)
    plot_vp(model, V, pi)
    plt.show()
```

# plot change in policy

```python
from cgitb import small
from typing import Callable
from matplotlib.pyplot import grid
from tqdm import tqdm
import numpy as np

from model import Model, Actions


def policy_iteration(model: Model, maxit: int = 100):
    V = np.zeros((model.num_states,))
    pi = np.zeros((model.num_states,))
    changes_per_iteration = []  # Step 1: Initialize the list to track changes

    def compute_value(s, a, reward: Callable):
        return np.sum(
            [
                model.transition_probability(s, s_, a)
                * (reward(s, a) + model.gamma * V[s_])
                for s_ in model.states
            ]
        )

    def policy_evaluation():
        for s in model.states:
            R = model.reward(s, pi[s])
            V[s] = compute_value(s, pi[s], lambda *_: R)

    def policy_improvement():
        changes = 0  # Initialize the number of changes for this iteration
        for s in model.states:
            action_values = [compute_value(s, a, model.reward) for a in Actions]
            best_action = np.argmax(action_values)
            if pi[s] != Actions(best_action):
                pi[s] = Actions(best_action)
                changes += 1  # Increment if there's a change in the policy
        return changes

    for i in tqdm(range(maxit)):
        for _ in range(5):
            policy_evaluation()
        pi_old = np.copy(pi)
        changes = policy_improvement()  # Get the number of changes
        changes_per_iteration.append(changes)  # Step 2: Record the changes
        if all(pi_old == pi):
            print("breaking")
            break

    return V, pi, changes_per_iteration  # Return the changes per iteration


if __name__ == "__main__":
    import matplotlib.pyplot as plt
    from world_config import cliff_world, small_world, grid_world
    from plot_vp import plot_vp

    model = Model(cliff_world)
    V, pi, changes = policy_iteration(model)  # Include changes in the return statement
    plot_vp(model, V, pi)

    # Step 3: Plot the changes
    plt.figure()
    plt.plot(changes)
    plt.xlabel('Iteration')
    plt.ylabel('Number of Changes in Policy')
#     plt.title('Policy Changes per Iteration')
#     plt.grid(True)
    plt.show()
```

```python
from cgitb import small
from typing import Callable
from matplotlib.pyplot import grid
from tqdm import tqdm
import numpy as np

from model import Model, Actions


def value_iteration(model: Model, maxit: int = 100):

    V = np.zeros((model.num_states,))
    pi = np.zeros((model.num_states,))

    def compute_value(s, a, reward: Callable):
        return np.sum(
            [
                model.transition_probability(s, s_, a)
                * (reward(s, a) + model.gamma * V[s_])
                for s_ in model.states
            ]
        )
    def policy_improvement():
        for s in model.states:
            action_index = np.argmax(
                [compute_value(s, a, model.reward) for a in Actions]
            )
            pi[s] = Actions(action_index)


    delta=1e+6
    delta_0=0.01

    while delta>=delta_0:

        delta=0
        for s in model.states:
            temp=V[s]
            vs = []
            for a in Actions:
                R = model.reward(s, a)
                v = compute_value(s, a, lambda *_: R)
                vs.append(v)
            V[s] = max(vs)
            delta = max(delta, abs(V[s] - temp))

    policy_improvement()

    return V, pi


if __name__ == "__main__":
    import matplotlib.pyplot as plt
    from world_config import cliff_world, small_world, grid_world
    from plot_vp import plot_vp

    model = Model(cliff_world)
    V, pi = value_iteration(model)
    plot_vp(model, V, pi)
    plt.show()
```

```python
from cgitb import small
from typing import Callable
from matplotlib.pyplot import grid
from tqdm import tqdm
import numpy as np

from model import Model, Actions

#max_iter prevent from going round and round

def sarsa(model: Model,max_iter: int = 20, n_episode: int = 1000, epsilon=0.01,alpha=0.3):
        
    Q = np.zeros((model.num_states, len(Actions)))

    for episode in range(n_episode):
        s = model.start_state
        coin = np.random.choice([0, 1], p=[1-epsilon, epsilon])
        if coin:
            a=np.random.randint(0, len(Actions))
        else:
            a= np.argmax(Q[s])
        
        for i in range(max_iter):
            
            
            s_=model.next_state(s, a)
            coin = np.random.choice([0, 1], p=[1-epsilon, epsilon])
            if coin:
                a_=np.random.randint(0, len(Actions))
            else:
                a_= np.argmax(
                    Q[s_]
                )
            Q[s][a]=Q[s][a]+alpha*(model.reward(s,a)+model.gamma*Q[s_,a_]-Q[s][a])
            s, a = s_, a_
            if s==model.goal_state:
                break
    V=np.max(Q,axis=1)
    pi=np.argmax(Q,axis=1)
    return V,pi


import matplotlib.pyplot as plt
from world_config import cliff_world, small_world, grid_world
from plot_vp import plot_vp

# Set global plot parameters for consistency
plt.rcParams['figure.figsize'] = [8, 6]  # Example size, adjust as needed
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

lr_study=[0,0.5,2.0]
for lr in lr_study:
    model = Model(small_world)
    V, pi = sarsa(model,alpha=lr)
    plot_vp(model, V, pi)
    plt.show()
```

```python
from cgitb import small
from typing import Callable
from matplotlib.pyplot import grid
from tqdm import tqdm
import numpy as np

from model import Model, Actions

#max_iter prevent from going round and round

def q_learning(model: Model,max_iter: int = 20, n_episode: int = 1000, epsilon=0.1, alpha=0.3, decay_alpha=False, decay_epsilon=False):
        
    Q = np.zeros((model.num_states, len(Actions)))
    iterations_per_episode = []  # Add this line

    for episode in range(n_episode):
        
        if decay_alpha:
            alpha=1/(episode+1)
            
        if decay_epsilon:
            epsilon=1/(episode+1)
        
        
        s = model.start_state
        coin = np.random.choice([0, 1], p=[1-epsilon, epsilon])
        if coin:
            a=np.random.randint(0, len(Actions))
        else:
            a= np.argmax(Q[s])
        iterations = 0  # Add this line
        for i in range(max_iter):
            iterations+=1
            
            
            s_ = model.next_state(s, a)
            coin = np.random.choice([0, 1], p=[1-epsilon, epsilon])
            if coin:
                a_=np.random.randint(0, len(Actions))
            else:
                a_= np.argmax(
                    Q[s_]
                )
            Q[s][a]=Q[s][a]+alpha*(model.reward(s,a)+model.gamma*np.max(Q[s_])-Q[s][a])
            s, a = s_, a_
            if s==model.goal_state:
                break
        iterations_per_episode.append(iterations)  # Add this line
    V=np.max(Q,axis=1)
    pi=np.argmax(Q,axis=1)

    return V,pi, iterations_per_episode


import matplotlib.pyplot as plt
from world_config import cliff_world, small_world, grid_world
from plot_vp import plot_vp

model = Model(small_world)
V, pi, temp = q_learning(model)
plot_vp(model, V, pi)
plt.show()
```

```python
def sarsa(model: Model, max_iter: int = 50, n_episode: int = 800, epsilon=0.01, alpha=0.3, dynamic_epsilon=False):
    Q = np.zeros((model.num_states, len(Actions)))
    rewards_per_episode = []  # Add this line

    for episode in range(n_episode):
        if dynamic_epsilon==True:
            epsilon=1/(episode+1)
        s = model.start_state
        coin = np.random.choice([0, 1], p=[1-epsilon, epsilon])
        if coin:
            a=np.random.randint(0, len(Actions))
        else:
            a= np.argmax(Q[s])
        rewards = 0  # Add this line
        
        for i in range(max_iter):
            rewards += model.reward(s,a)  # Add this line
            
            s_=model.next_state(s, a)

            coin = np.random.choice([0, 1], p=[1-epsilon, epsilon])
            if coin:
                a_=np.random.randint(0, len(Actions))
            else:
                a_= np.argmax(Q[s_])
            Q[s][a]=Q[s][a]+alpha*(model.reward(s,a)+model.gamma*Q[s_,a_]-Q[s][a])
            s, a = s_, a_
            if s==model.goal_state:
                break
        rewards_per_episode.append(rewards)  # Add this line

    V = np.max(Q, axis=1)
    pi = np.argmax(Q, axis=1)
    return V, pi, np.array(rewards_per_episode)  # Modify this line
```

```python
from scipy.ndimage.filters import uniform_filter1d

def plot_cum_rewards(cum_r_sarsa, cum_r_ql, epsilon=0.1, min_val=None):
    y_sarsa = uniform_filter1d(cum_r_sarsa, size=40)
    y_ql = uniform_filter1d(cum_r_ql, size=40)

    plt.plot(range(len(y_sarsa)), y_sarsa, label='SARSA')
    plt.plot(range(len(y_ql)), y_ql, label='Q Learning')
    if not min_val:
        min_val = max(np.min(y_sarsa), np.min(y_ql))
    plt.ylim(min_val, 10)
    plt.xlabel("Number of episode")
    plt.ylabel("Cumulative reward")
    plt.title("Rewards obtained in each episode - (" + r"$\epsilon = "+str(epsilon)+"$)")
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from world_config import cliff_world, small_world, grid_world
from plot_vp import plot_vp

epsilon_val=0.3

model = Model(cliff_world)
V, pi, cumR_ql = q_learning(model, max_iter = 100, n_episode= 2000, epsilon=epsilon_val)

V, pi, cumR_sarsa = sarsa(model, max_iter = 100, n_episode= 2000, epsilon=epsilon_val)

cum_sarsa = cumR_sarsa
cum_ql = cumR_ql
NREPS=10

for i in range(NREPS):
    
    V_sarsa, pi_sarsa, cumR_sarsa  = sarsa(model, max_iter = 100, n_episode= 2000, epsilon=epsilon_val)
    V_ql, pi_ql, cumR_ql =q_learning(model, max_iter = 100, n_episode=2000, epsilon=epsilon_val)
    cum_sarsa += cumR_sarsa
    cum_ql += cumR_ql

cum_sarsa /= (NREPS+1)
cum_ql /= (NREPS+1)


    
plot_cum_rewards(cum_sarsa,cum_ql,epsilon=epsilon_val, min_val=-500)

```
